```python
df = pd.read_csv('nilaisen.csv')

# Handle missing values (NaN) in 'kata' column
df['tweet'].fillna('', inplace=True)

# Filter rows that contain the word "tidak" in the 'kata' column
tidak_tweets = df[df['tweet'].str.contains(r'\bprabowo\b', case=False, na=False)]

# Print the tweets that contain the word "tidak" in the 'kata' column
print("Tweets that contain the word 'tidak':")
for tweet in tidak_tweets['tweet']:
    print(tweet)

# Count the number of tweets that contain the word "tidak" for each sentiment category
tidak_sentiment_counts = tidak_tweets['sentiment'].value_counts()

# Print the counts for each sentiment category
print("\nSentiment counts for tweets that contain the word 'tidak':")
print(tidak_sentiment_counts)
```
